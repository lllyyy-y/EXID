```python
# 2. 라이브러리 불러오기
import pandas as pd
from sklearn.model_selection import train_test_split
from flaml import AutoML
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# 3. 데이터 불러오기 (Colab 업로드 방식)
from google.colab import files
uploaded = files.upload()   # 'ObesityDataSet_raw_and_data_sinthetic.csv' 선택

# 4. CSV 파일 로드
df = pd.read_csv("ObesityDataSet_raw_and_data_sinthetic.csv")

# 5. X, y 분리
X = df.drop(columns=["NObeyesdad", "Height", "Weight"])
y = df["NObeyesdad"]

# 6. 학습/테스트 데이터 분할 (75:25, stratify=y로 클래스 균형 유지)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
# 7. AutoML 설정
automl = AutoML()
settings = {
    "time_budget": 60,  # 총 탐색 시간(초). 필요 시 조정 가능
    "metric": "accuracy",  # 평가 지표
    "task": "classification",  # 분류 문제
    "log_file_name": "flaml_obesity.log",
    "seed": 42
}

# 8. AutoML 훈련
automl.fit(X_train=X_train, y_train=y_train, **settings)
```

```python
print(automl.best_config_per_estimator)
```

```python
from sklearn.metrics import accuracy_score

results = {}

for estimator, config in automl.best_config_per_estimator.items():
    try:
        # 같은 데이터에서 개별 알고리즘 재훈련
        model = AutoML()
        model.fit(
            X_train=X_train, y_train=y_train,
            estimator_list=[estimator],  # 특정 알고리즘만 사용
            time_budget=60,              # 1분만
            metric="accuracy",
            seed=42
        )
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred)
        results[estimator] = acc
    except Exception as e:
        print(estimator, "error:", e)
```

```python
print("모델별 정확도:", results)
```
